```python
# Remove $ and , from total_paid
df['total_paid'] = df['total_paid'].str.replace('$', '')
df['total_paid'] = df['total_paid'].str.replace(',', '')
df
```

```python
# Remove rows that have $0 sales
df.drop(df.loc[df['total_paid']== "0"].index, inplace=True)
df
```
